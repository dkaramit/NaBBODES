```python
import subprocess
import sys

import os

import time
```

```python
import numpy as np

%matplotlib widget
import matplotlib.pyplot as plt
#-------------------------#

from scipy.integrate import odeint
from scipy.integrate import Radau
```

```python
os.chdir('.')
os.system(r'make Robertson.run')
```

```python
time0=time.time()

output=subprocess.check_output(["./Robertson.run"]).decode(sys.stdout.encoding).split("\n")

print("time: {:10} s".format( time.time()-time0)  )

solution=np.array([ (i.split(' '))[:-1]   for i in output[:-1] ] ,np.float64)

```

```python
t=solution[:,0]
y1=solution[:,1]
y2=solution[:,2]
y3=solution[:,3]

err1=solution[:,4]
err2=solution[:,5]
err3=solution[:,6]
```

```python
t[-1]/1e5
```

```python
def f(t,y):
    lhs=np.zeros(3)
    lhs[0]= -0.04*y[0]+1e4*y[1]*y[2] ;
    lhs[1]= 0.04*y[0]-1e4*y[1]*y[2]-3e7* y[1]*y[1] ;
    lhs[2]=  3e7*y[1]*y[1];
    return lhs
```

```python
sol_py=Radau(f,0,[y1[0],y2[0],y3[0]],t[-1],atol=1e-8,rtol=1e-8)

time0=time.time()

y_py=[[y1[0],y2[0],y3[0]]]
t_py=[0]
while sol_py.status=='running' :
    sol_py.step()
    y_py.append(sol_py.y)
    t_py.append(sol_py.t)
#     print(sol_py.step_size,sol_py.t)
y_py=np.array(y_py)

print("time: {:10} s".format( time.time()-time0)  )
```

```python
def g(y,t):
    return f(t,y)
```

```python
time0=time.time()
sol_ode=odeint(g,y_py[0],t )
print("time: {:10} s".format( time.time()-time0)  )
```
